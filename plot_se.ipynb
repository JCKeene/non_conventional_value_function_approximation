```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

```python
environment = "cartpole"

models = ["Neural Network", "Linear Model", "Decision Tree", "Random Forest", "K-Neighbours", "Support Vectors", "Gaussian Process (FQI)", "Gaussian Process (Online)"]
legends = ["Neural Network", "Linear Model", "Decision Tree (FQI)", "Random Forest (FQI)", "K-Neighbours (FQI)", "Support Vectors (FQI)", "Gaussian Process (FQI)", "Gaussian Process (Online)"]
n = len(legends)

n_eps = []
n_steps = []
not_solved = []

for i in range(len(legends)):
    a = np.genfromtxt(f'results/{environment}/se/{environment}_se195_{models[i]}.csv', delimiter=',')
    n_eps.append(a[0,:])
    n_steps.append(a[1,:])
    not_solved.append(a[2,:])
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax = sns.boxplot(data=n_steps, orient="h", palette="Set2")
ax.set_yticklabels(legends)
ax.set_xlabel('Time-steps')

fig.tight_layout()
plt.savefig("cartpole_se195.pdf", format="pdf", bbox_inches="tight")
```
